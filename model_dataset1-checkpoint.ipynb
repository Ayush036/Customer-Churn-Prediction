```python
import pandas as pd
dataset=pd.read_csv("telecom.csv")
dataset["TotalCharges"]=pd.to_numeric(dataset["TotalCharges"],errors="coerce")
a=dataset["TotalCharges"].dropna()
avg=np.average(a)
dataset["TotalCharges"]=dataset["TotalCharges"].fillna(avg)
#dataset["TotalCharges"]=dataset["TotalCharges"].dropna()
pd.isnull(dataset["TotalCharges"]).sum()
dataset["TotalCharges"]=dataset["TotalCharges"].astype("float64")
dataset.drop(labels=["customerID"],axis=1,inplace=True)
dataset=pd.get_dummies(data=dataset,columns=["gender",'Partner','SeniorCitizen','PhoneService','MultipleLines','Dependents','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling','PaymentMethod','Churn'],drop_first=True)
```

```python
y=dataset["Churn_Yes"].values
x=dataset.drop(labels=["Churn_Yes"],axis=1)



#x=dataset.iloc[:,:-1]
#y=dataset.iloc[dataset['Churn_yes']]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
#from sklearn.metrics import confusion_matrix
#confusion_matrix=confusion_matrix(y_test,y_pred)
round(logreg.score(x_test,y_test),ndigits=3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
from sklearn.metrics import accuracy_score
y_pred=knn.predict(x_test)
accuracy_score(y_test,y_pred)

```
